```python
# transforming this data 
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("'", "")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

```python
df.columns
```

```python
drop_cols = [
    'external_ids',
    'associated_types',
    'aid_contribution_000_us$',
    'river_basin',
    'reconstruction_costs_000_us$',
    'reconstruction_costs,_adjusted_000_us$',
    'insured_damage_000_us$',
    'insured_damage,_adjusted_000_us$'
]

df.drop(columns=drop_cols, inplace=True, errors='ignore')
```

```python
cat_fill_cols = [
    'origin',
    'magnitude_scale',
    'location'
]

df[cat_fill_cols] = df[cat_fill_cols].fillna("Unknown")
```

```python
df['magnitude'] = df['magnitude'].fillna(df['magnitude'].median())
df['latitude']  = df['latitude'].fillna(df['latitude'].median())
df['longitude'] = df['longitude'].fillna(df['longitude'].median())
```

```python
impact_cols = [
    'total_deaths',
    'no._injured',
    'no._affected',
    'no._homeless',
    'total_affected'
]

df[impact_cols] = df[impact_cols].fillna(0)
```

```python
damage_cols = [
    'total_damage_000_us$',
    'total_damage,_adjusted_000_us$'
]

df[damage_cols] = df[damage_cols].fillna(0)
```

```python
df.isnull().sum()
```

```python
events = df[[
    'disno.',
    'disaster_group',
    'disaster_type',
    'disaster_subtype',
    'event_name',
    'country',
    'region',
    'subregion',
    'location',
    'start_year',
    'start_month',
    'start_day',
    'end_year',
    'end_month',
    'end_day'
]]

events.to_csv("disaster_events.csv", index=False)
```
